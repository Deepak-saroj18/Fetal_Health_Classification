#### Destribution Analysis :

```python
# Plotiing histograms and box plots for each feature to visualize their distrubution

# Set up the matplotlib figure
plt.figure(figsize=(20,40))

# Loop through each column in the dataset
for i, column in enumerate(df.columns):
    # plot histogram
    plt.subplot(22, 2, 2*i+1) # adjust grid layout for histograms
    sns.histplot(df[column], kde=True) # Plot histogram with KDE
    plt.title(f"Histogram of {column}")

    # plot boxplot
    plt.subplot(22,2,2*i+2)
    sns.boxplot(y=df[column])
    plt.title(f"Boxplot of {column}")
    plt.xlabel('')
    
plt.tight_layout()
plt.show()
```
